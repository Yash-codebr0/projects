```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt


# Get Data

def get_stock_data(ticker):
    df = yf.download(ticker, start="2015-01-01", auto_adjust=True)

    if df.empty:
        return None

    return df[["Open", "High", "Low", "Close", "Volume"]]


# Features

def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA5'] = df['Close'].rolling(5).mean()
    df['MA10'] = df['Close'].rolling(10).mean()
    df['MA20'] = df['Close'].rolling(20).mean()
    df['MA50'] = df['Close'].rolling(50).mean()
    df['MA120'] = df['Close'].rolling(120).mean()
    df['MA365'] = df['Close'].rolling(365).mean()
    
    # Volatility
    df['Volatility'] = df['Return'].rolling(50).std()
    df['Volume_Change'] = df['Volume'].pct_change()
    
    # RSI (simple)
    delta = df['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    rs = gain.rolling(14).mean() / loss.rolling(14).mean()
    df['RSI'] = 100 - (100 / (1 + rs))

    # MACD
    ema12 = df['Close'].ewm(span=12).mean()
    ema26 = df['Close'].ewm(span=26).mean()
    df['MACD'] = ema12 - ema26
    df['Signal'] = df['MACD'].ewm(span=9).mean()

    # Momentum
    df['Momentum'] = df['Close'] - df['Close'].shift(365)

    # Target 
    df['Future_Return'] = df['Close'].shift(-5) / df['Close'] - 1
    df['Target'] = np.where(df['Future_Return'] > 0.02, 1, 0) 

    return df.dropna()


# Data preparing

def prepare_data(df):
    features = ["Open", "High", "Low", "Close",
                "MA5", "MA10", "MA20", "MA50","MA120",
                "MA365","Momentum","RSI", "MACD",
                "Signal", "Volatility",
                "Volume"]

    X = df[features]
    y = df['Target']  

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    split = int(len(X_scaled) * 0.8)
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y[:split], y[split:]

    return X_train, X_test, y_train, y_test, scaler


# train Model

def train_model(X_train, y_train):
    model = XGBClassifier(n_estimators=500, max_depth=6,
                          learning_rate=0.05,
                          subsample=0.8,colsample_bytree=0.8,
                          random_state=42)
    model.fit(X_train, y_train)
    return model


# Evaluation

def eval(model, X_test, y_test):
    y_pred =model.predict(X_test)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    return y_pred

    
# Plot visualization
def plot_result(y_test,y_pred, ticker):
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.reset_index(drop=True), label='Actual', alpha=0.7)
    plt.plot(pd.Series(y_pred), label='Predicted', alpha=0.7)
    plt.title(f"{ticker} Stock Trend Prediction")
    plt.legend()
    plt.show()

# Run

def run(ticker):
    print(f"\nProcessing {ticker} data...")

    df = get_stock_data(ticker)
    if df is None or df.empty:
        print(f"Skipping {ticker} (no data)")
            
    df = add_features(df)
    if len(df) < 100:
        print(f"Skipping {ticker} (not enough data)")
            
    
    X_train, X_test, y_train, y_test,scaler = prepare_data(df)
    model = train_model(X_train, y_train)
    y_pred = eval(model, X_test, y_test)
    plot_result(y_test, y_pred, ticker)
    
    return model,scaler



model,scaler=run("AAPL")
```

```python
import joblib
joblib.dump(model, "stp.pkl")
print("Model saved successfully")
```
